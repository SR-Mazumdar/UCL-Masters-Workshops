### Main verbs

* Head and tail
* select (and unselect)
* mutate (new columns)
* Filter, arrange, and unique

Note that for the below commands, we specify specific columns via backticks. This is because R does not like variable names that start with numbers.

```r
play.table = tab %>% head(n=101) %>% tail(n=50) %>% 
    ## Here we select specific columns
    select(c(gene, "MUT_2-1", "WT_2-1")) %>%
    ## Here we create new columns
    mutate("TEST_2" = 1) %>%
    mutate("sum_of_2" = `MUT_2-1` +  `WT_2-1`) %>%
    ## Here we filter
    filter("sum_of_2" != 0) %>% filter("sum_of_2" > 1) %>%
    ## Here we will sort (in descending)
    arrange(desc("MUT_2-1"))
```

```r
play.table
```

Let's produce a column that shows the minimum value of the columns 2_MUT-1 or 2_WT-1. 

```r
play.table %>% mutate("myrowmin" = min(`MUT_2-1`, `WT_2-1`)) 
```

Uh-oh. Something wen't wrong. We're getting 0 for every row. Why?

* The tidyverse assumes that every operation you do is being applied "column-wise". So when you ask it to calculate the minimum of column X and Y via `min(X,Y)` it's not doing this on a row-by-row basis, but it takes the entire list (i.e. `min(188.99, 56.21, 47.49, ..., 0.00, 255.24, 107.21, ...)`)

We can tell it to stop doing that, by specifying that it should use `rowwise` operations instead.

### Extra verbs

* Rowwise operations:
 * Mean, median, max, min
* Ungroup

```r
play.table = play.table %>% rowwise %>%
    mutate("myrowmin" = min(c(`MUT_2-1`, `WT_2-1`))) %>%         ## row-wise min
    mutate("myrowmax" = max(c(`MUT_2-1`, `WT_2-1`))) %>%         ## row-wise max
    mutate("myrowmedian" = median(c(`MUT_2-1`, `WT_2-1`))) %>%   ## row-wise median
    mutate("myrowmean" = mean(c(`MUT_2-1`, `WT_2-1`))) %>%       ## row-wise mean
    mutate("myrowsd" = sd(c(`MUT_2-1`, `WT_2-1`)))               ## row-wise standard-deviation
```

Note that at this point, each of the values in a single row are grouped! We can undo this by using the `ungroup` operation to revert to the normal state of column-wise operations.

```r
play.table %>% 
    ungroup() %>%                         ## this undos rowwise operations
    select(-c("TEST_2", "sum_of_2"))
```

### Advanced verbs

* Pivot (long or wide)
* Separate (**not**: "seperate")
* Group and Nest

```r
tab %>% tail()
```

Above we have the first 6 lines of our table. Note that this is what is known as `wide format`. Why?

* A single row contains multiple "observations" (2_WT-1, 2_WT-2, etc.)

We can change this, so that a single observation exists on a single row. To do this, we need to `pivot` the table so that one of the columns becomes FIXED ("gene"), but the other columns are encoded directly into the table as row data.

This is known as pivotting from "wide" to "long"

```r
tab %>%                                                ## Originally in Wide Format
    ## By default, pivot_longer tries to encode all columns
    ## we can tell it to use "gene" as the FIXED column, by DESELECTING it via `-c`
    pivot_longer(-c(gene)) %>%                         ## Convert to Long Format
    head(n=12)
```

Note that now, we have a new column called "name" that encodes the columns we had before, and that a single gene is specified 8 times, with different "name" and "value" values. This is long  format.

Note also that we can pivot back from "long" to "wide" with an equivalent command `pivot_wider`:

```r
tab %>%                                                   ## Originally in Wide Format
    pivot_longer(-c(gene)) %>%                            ## Convert to Long Format
    ## Here we tell it to use the values from the "name" column, and convert
    ## these to columns. The values in each of these columns will come from
    ## the "value" column.
    pivot_wider(names_from=name, values_from=value) %>%   ## Convert back to Wide Format
    head()
```

The above table is our Replicate matrix


# AIM: Replicate Matrix → Sample Matrix

Collapse all of these values into an Average matrix. 

Ideally, we shouldn't have 8 value columns of 2_WT-1, 2_WT-2, 2.5_WT-1, 2.5_WT-2, etc., but 4 columns of 2_WT, 2.5_WT, 2_MUT, etc.

**We can accomplish this via Pivot**

```r
tab %>%                                             ## Originally in Wide Format
    pivot_longer(-c(gene)) %>% tail()
```

We are now going to "sep**a**rate" (notice the "a"), the "name" column into two columns, by splitting the text of the "-" character, so that "2_WT-1" becomes "2_WT" and "1", each filling a space in a column named "sample" and the other in "replicate_number".

```r
tab %>%                                             ## Originally in Wide Format
    pivot_longer(-c(gene)) %>%                      ## Convert to Long Format
    separate(name, into=c("sample", "replicate_number"), sep="-") %>%
    head()
```

Nice, so now we have created two new columns "sample" and "replicate_number" based on a previous column "name" that no longer exists.

```r
##  Let's remove the replicate number, we don't need it
long.tab = tab %>%
    pivot_longer(-c(gene)) %>%
    separate(name, into=c("sample", "replicate_number"), sep="-") %>%
    select(-c("replicate_number"))
long.tab %>% tail()
```

So we for each unique sample we now have two values for the same gene. Let's try to **summarize** this information so that a single gene has a single value for a given sample.

* First, we "group" the columns we want to summarize. 
  
  We want unique gene and sample values, so these are the two columns we will **group_by**.

* Second, we will "summarize" these grouped values by applying a function to the columns that were not specified in the group_by command ("value")

Let's collapse the values that share the same gene and the same sample, and summarize them into an average value.

#### Nice Summarization

```r
## "Nice" way
long.tab %>%
    group_by(gene, sample) %>%
    ## the first "mean" is the name of column
    ## the second "mean" used here, is an in-built function in R
    summarize(mean = mean(value)) %>% tail()
```

## Average long table to Average Matrix

Let's use the "Nice" summarization method to calculate the average value of each sample from the replicates, and then convert it back to a wide matrix.

```r
## Convert this back to a regular wide matrix
avg.matrix = long.tab %>%
    group_by(gene, sample) %>%
    summarise(mean = mean(value)) %>%
    ungroup(c(gene,sample)) %>% 
    pivot_wider(names_from=c("sample"), 
                values_from=c("mean"))
avg.matrix %>% head()
```

```r
# the "nice" summarization table
long.tab %>%
    group_by(gene, sample) %>%
    summarise(mean = mean(value)) %>%
    ungroup(c(gene,sample)) %>% head()
```

```r
## Convert this back to a regular wide matrix
long.tab %>%
    group_by(gene, sample) %>%
    summarise(mean = mean(value)) %>%
    ungroup(c(gene,sample)) %>% 
    ## We pivot from long to wide, populating columns with names from "sample",
    ## and the values from "mean"
    pivot_wider(names_from=c("sample"), 
                values_from=c("mean")) %>% head()
```

Nice! Okay, let's store this into a variable called "avg.matrix" so we can use it later.

```r
# Above command
avg.matrix = long.tab %>%
    group_by(gene, sample) %>%
    summarise(mean = mean(value)) %>%
    ungroup(c(gene,sample)) %>% 
    pivot_wider(names_from=c("sample"), 
                values_from=c("mean"))
avg.matrix %>% head()
```

One extra thing we can do is convert all the values into integers.

```r
new_mat = as.matrix(avg.matrix %>% column_to_rownames("gene"))
int_mat = apply(new_mat, c(1,2), as.integer)
int_mat %>% head()
```
